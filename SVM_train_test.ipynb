Setting the dataset

```python
X_train = None
y_train = None
X_test = None
y_test = None
```

Building the model

```python
from sklearn import svm

model = svm.LinearSVC()
```

Training

```python
from time import time
import datetime

start_time = time()

model.fit(X_train, y_train)

end_time = time()
duration = end_time-start_time

print("Training duration %d" %(str(datetime.timedelta(seconds=duration))))
```

Get the support vectors

```python
support_vectors = model.support_vectors_
dec = model.decision_function([[1]])
```

Test

```python
from sklearn.metrics import classification_report

print("TEST RESULTS")
y_test_model = model.predict(X_test)
print(classification_report(y_test, y_test_model))
```

Save : https://scikit-learn.org/stable/model_persistence.html

```python
from joblib import dump

dump(model, 'model_svm.joblib')

# later, the model can be retrieved with "clf = load('filename.joblib')"
```
